```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.drop(['PassengerId', 'Name',], axis = 1)
df.head()
```

# Looking for null values 

```python
df.isnull().sum()
```

# Handeling Null values 

```python
df[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] = df[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(value=0)
df.isnull().sum().sort_values(ascending=False)
df.isnull().sum()
```

# Convert boolean vals to int

```python
df['Transported'] = df['Transported'].astype(int)
df['VIP'] = df['VIP'].astype(int)
df['CryoSleep'] = df['CryoSleep'].astype(int)
df.head() 
```

# Splitting Column 'Cabin' into 3 sub columns for the sake pf ease in model training 

```python
df[['Deck', 'Cabin_num', 'Side']] = df['Cabin'].str.split('/', expand = True)

# moving Transported Column to the end
column_to_move = df.pop('Transported')
df.insert(14, 'Transported', column_to_move)

df.head()
```

```python
try:
        df = df.drop('Cabin', axis=1)
except KeyError:
        print('Column Not Found')
df.head()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import  SimpleImputer

le = LabelEncoder()
for column_name in df.columns:
    if df[column_name].dtype == object:
        df[column_name] = le.fit_transform(df[column_name])
    else:
        pass
df.head()
```

# Dataset split

```python
def split_dataset(dataset, test_ratio=0.20):
        imputer = SimpleImputer(strategy="most_frequent")
        dataset = pd.DataFrame(imputer.fit_transform(dataset))     
        test_indices = np.random.rand(len(dataset)) < test_ratio
        train = dataset[~test_indices] 
        valid = dataset[test_indices]
        return train[train.columns[:-1]].values, train[train.columns[-1]].values, valid[valid.columns[:-1]].values, valid[valid.columns[-1]].values

X_train, y_train, X_valid, y_valid  = split_dataset(df)
print(len(X_train), len(y_train), len(X_valid), len(y_valid))
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import  classification_report
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

# Validation

```python
y_pred  = clf.predict(X_valid)
print(classification_report(y_valid, y_pred))
```

# Test Submission

```python
test_df = pd.read_csv('test.csv')
submission_id = test_df.PassengerId
test_df = test_df.drop(['PassengerId', 'Name',], axis = 1)
# Replace NaN values with zero
test_df[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] = test_df[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(value=0)
test_df.isnull().sum()

# Creating new features Deck, Cabin_num, Side from column Cabin and remove Cabin 
test_df[['Deck', 'Cabin_num', 'Side']] = test_df['Cabin'].str.split('/', expand = True)
test_df = test_df.drop('Cabin', axis=1)




# Convert boolean into int 
test_df['VIP'] = test_df['VIP'].astype(int)
test_df['CryoSleep'] = test_df['CryoSleep'].astype(int)

# Encoding Object Values
for column_name in test_df.columns:
    if test_df[column_name].dtype == object:
        test_df[column_name] = le.fit_transform(test_df[column_name])
    else:
        pass

# impute NaN values 
imputer = SimpleImputer(strategy="most_frequent")
test_df = pd.DataFrame(imputer.fit_transform(test_df)) 

# make predictions
predictions = clf.predict(test_df)


output = pd.DataFrame({'PassengerId': submission_id , 'Transported': predictions})
output['Transported'] = output['Transported'].astype(bool)

output.to_csv('titanic_spaceship.csv', index = False)
output.head()

```
